R -AC Saf Ses Ortalaması(Db) ve L-AC Saf Ses Ortalaması(Db) 16-25 dB arasında ise Çok Hafif Dereceli İşitme Kaybı olarak adlandırılır.Bunun dağılımını ve cinsiyetlere göre dağılımını inceleyelim.

```python
cok_hafif_ik = df[(df['R_Safses_ortalama'] >= 16) & (df['R_Safses_ortalama'] <= 25) &
                       (df['L_Safses_ortalama'] >= 16) & (df['L_Safses_ortalama'] <= 25)]

kisi_sayisi = cok_hafif_ik.shape[0]
print("Çok Hafif dereceli işitme kaybı olan kişi sayısı:", kisi_sayisi)
```

```python
plt.figure(figsize=(4, 3)) 
plt.hist(cok_hafif_ik['CİNSİYET'], bins=10, color='orange', alpha=0.7)
plt.xlabel('CİNSİYET')
plt.ylabel('Kişi sayısı')
plt.title("Çok Hafif Dereceli İşitme Kaybının Cinsiyet Dağılımı")
plt.grid(True)
plt.show()
```
